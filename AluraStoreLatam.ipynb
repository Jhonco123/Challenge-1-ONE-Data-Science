### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt


url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

#1. Análisis de facturación



```python
# Guardar las bases de datos en un diccionario
datos_tiendas = {
    "Tienda 1": tienda1,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4
}

# Función para calcular el ingreso total de una tienda
def calcular_ingreso_total(df):
    return df["Precio"].sum()

# Calcular ingresos y guardarlos en una lista
ingresos = []
nombres_tiendas = []

for nombre, i in datos_tiendas.items():
    ingreso = calcular_ingreso_total(i)
    ingresos.append(ingreso)
    nombres_tiendas.append(nombre)

# Crear diccionario final con resultados
ingresos_dict = dict(zip(nombres_tiendas, ingresos))

# Encontrar la tienda con menor ingreso
min_ingreso = min(ingresos)
tienda_menor = ""

for nombre, ingreso in ingresos_dict.items():
    if ingreso == min_ingreso:
        tienda_menor = nombre
        break

# Imprimir resultados
print("Ingresos Totales por Tienda:")
for nombre, ingreso in ingresos_dict.items():
    print(f"{nombre}: ${ingreso:,.0f}")

print(f"\n La Tienda con menor ingreso es: {tienda_menor}")

# Gráfica de barras
plt.figure(figsize=(8, 5))
barras = plt.bar(nombres_tiendas, ingresos, color=["#4CAF50", "#2196F3", "#FFC107", "#F44336"])
plt.title("Ingresos Totales por Tienda")
plt.xlabel("Tienda")
plt.ylabel("Ingreso (COP)")

# Etiquetas sobre las barras
for bar, ingreso in zip(barras, ingresos):
    plt.text(bar.get_x() + bar.get_width()/2, ingreso + 1_000_000, f"${int(ingreso):,}", ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
# Guardar las bases de datos en un diccionario
datos_tiendas = {
    "Tienda 1": tienda1,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4
}

# Limpiar columnas
for df in datos_tiendas.values():
    df.columns = df.columns.str.strip()

# Función para obtener top 3 de categorías
def top_categorias(df):
    if "Categoría del Producto" in df.columns:
        return df["Categoría del Producto"].value_counts().head(3)
    else:
        return pd.Series(dtype=int)  # Vacía si no existe la columna

# Construir la tabla de resultados
tabla_top = []
categorias_globales = set()

for nombre, df in datos_tiendas.items():
    top = top_categorias(df)
    categorias_globales.update(top.index)
    tabla_top.append({
        "Tienda": nombre,
        "Top 1": f"{top.index[0]} ({top.iloc[0]})" if len(top) > 0 else "N/A",
        "Top 2": f"{top.index[1]} ({top.iloc[1]})" if len(top) > 1 else "N/A",
        "Top 3": f"{top.index[2]} ({top.iloc[2]})" if len(top) > 2 else "N/A"
    })

# Mostrar tabla
df_top_categorias = pd.DataFrame(tabla_top)
print("📊 Top 3 Categorías Más Vendidas por Tienda:\n")
print(df_top_categorias.to_string(index=False))

# Estructura para conteo de cada categoría por tienda
conteo_por_categoria = {cat: [] for cat in categorias_globales}
tiendas = list(datos_tiendas.keys())

for cat in categorias_globales:
    for tienda in tiendas:
        df = datos_tiendas[tienda]
        conteo = df["Categoría del Producto"].value_counts().get(cat, 0)
        conteo_por_categoria[cat].append(conteo)

# Crear gráfica de barras agrupadas

categorias_graficar = ['Muebles', 'Electrónicos', 'Juguetes']
colores = ['#2980b9', '#27ae60', '#f39c12']
anchura_barra = 0.2

# Posiciones base
tiendas = list(datos_tiendas.keys())
posiciones = list(range(len(tiendas)))

# Crear la figura
plt.figure(figsize=(10, 6))

# Graficar barras por categoría
for i, categoria in enumerate(categorias_graficar):
    valores = []
    for tienda in tiendas:
        df = datos_tiendas[tienda]
        conteo = df["Categoría del Producto"].value_counts().get(categoria, 0)
        valores.append(conteo)

    # Ajuste horizontal por categoría
    posiciones_categoria = [x + i * anchura_barra for x in posiciones]
    plt.bar(posiciones_categoria, valores, width=anchura_barra, label=categoria, color=colores[i])

# Ajustar etiquetas y leyenda
plt.xticks([p + anchura_barra for p in posiciones], tiendas)
plt.xlabel("Tiendas")
plt.ylabel("Cantidad Vendida")
plt.title("Top 3 Categorías Más Vendidas por Tienda")
plt.legend()
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
# Función para calcular calificación promedio
def calcular_calificacion_promedio(df, nombre_tienda):
    if "Calificación" in df.columns:
        promedio = round(df["Calificación"].mean(), 2)
        return promedio
    else:
        print(f"Error: La columna 'Calificación' no existe en {nombre_tienda}")
        return 0

# Crear lista de tiendas y diccionario con resultados
lista_tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
dataframes = [tienda1, tienda2, tienda3, tienda4]
calificaciones_dict = {}

for nombre, df in zip(lista_tiendas, dataframes):
    calificacion = calcular_calificacion_promedio(df, nombre)
    calificaciones_dict[nombre] = calificacion

# Crear tabla para mostrar resultados
calificaciones = pd.DataFrame.from_dict(calificaciones_dict, orient='index', columns=["Calificación Promedio"])
calificaciones.sort_values(by="Calificación Promedio", ascending=False, inplace=True)

# Mostrar resultados y aplicar condición
print("Calificación Promedio por Tienda:\n")
print(calificaciones)

mejor_tienda = calificaciones.index[0]
print(f"\n La tienda que tiene mayor satisfacción es: {mejor_tienda}")

# Gráfica tipo pastel con matplotlib
plt.figure(figsize=(7, 7))
plt.pie(
    calificaciones["Calificación Promedio"],
    labels= calificaciones.index,
    autopct="%1.2f%%",
    startangle=90,
    colors=["#ff9999", "#66b3ff", "#99ff99", "#ffcc99"]
)
plt.title("Distribución de la Calificación Promedio por Tienda")
plt.axis("equal")
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python

# Función para obtener productos extremos
def obtener_extremos(df):
    if "Producto" not in df.columns:
        return ("Sin datos", 0), ("Sin datos", 0)
    conteo = df["Producto"].value_counts()
    return (conteo.idxmax(), conteo.max()), (conteo.idxmin(), conteo.min())

#  Cargar datos
tiendas = {
    "Tienda 1": tienda1,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4
}

# Aplicar la función y almacenar resultados
mas_vendidos = {}
menos_vendidos = {}

for nombre, df in tiendas.items():
    mas, menos = obtener_extremos(df)
    mas_vendidos[nombre] = mas
    menos_vendidos[nombre] = menos

# Mostrar tabla
tabla = pd.DataFrame({
    "Más Vendido": {k: f"{v[0]} ({v[1]})" for k, v in mas_vendidos.items()},
    "Menos Vendido": {k: f"{v[0]} ({v[1]})" for k, v in menos_vendidos.items()}
})

print(" Productos Más y Menos Vendidos por Tienda:\n")
print(tabla)

#  gráfico
labels = list(tiendas.keys())
mas_cant = [v[1] for v in mas_vendidos.values()]
menos_cant = [v[1] for v in menos_vendidos.values()]

x = range(len(labels))
width = 0.35

plt.figure(figsize=(10, 6))
plt.bar(x, mas_cant, width=width, label='Más Vendido', color="#4CAF50")
plt.bar([p + width for p in x], menos_cant, width=width, label='Menos Vendido', color="#F44336")

plt.xlabel("Tiendas")
plt.ylabel("Cantidad Vendida")
plt.title("Comparación de Productos Más y Menos Vendidos por Tienda")
plt.xticks([p + width / 2 for p in x], labels)
plt.legend()
plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
# Crear diccionario con los DataFrames

tiendas = {
    "Tienda 1": tienda1,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4
}

# Limpiar nombres de columnas
def limpiar_columnas(df):
    df.columns = df.columns.str.strip().str.lower()
    return df

tienda1 = limpiar_columnas(tienda1)
tienda2 = limpiar_columnas(tienda2)
tienda3 = limpiar_columnas(tienda3)
tienda4 = limpiar_columnas(tienda4)

# Función para calcular el costo promedio de envío

def calcular_promedio_envio(df):
    columnas_posibles = ["costo de envío", "costo de envio"]
    for col in columnas_posibles:
        if col in df.columns:
            return round(df[col].mean())
    return None

#  Calcular promedios
costos_envio = []
nombres_tiendas = []

for nombre, df in tiendas.items():
    promedio = calcular_promedio_envio(df)
    if promedio is not None:
        costos_envio.append(promedio)
        nombres_tiendas.append(nombre)

#  Mostrar resultados
print("Costo Promedio de Envío por Tienda (COP):\n")
for tienda, costo in zip(nombres_tiendas, costos_envio):
    print(f"{tienda}: ${costo:,}")

# Gráfica
plt.figure(figsize=(8, 5))
bars = plt.bar(nombres_tiendas, costos_envio, color="#2ecc71")
plt.title("Costo Promedio de Envío por Tienda")
plt.xlabel("Tienda")
plt.ylabel("Costo (COP)")

# Etiquetas encima de las barras
for bar, costo in zip(bars, costos_envio):
    plt.text(bar.get_x() + bar.get_width()/2, costo + 300, f"${costo:,}", ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
